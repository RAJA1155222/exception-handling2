Q4. Why LookupError class is used?

Explain with KeyError and IndexError.

LookupError is the base class for errors that occur when we try to access
an invalid key or index in a collection.

It mainly covers:

KeyError

IndexError

Example of KeyError

```python
data = {"name": "Raja"}

print(data["age"])
```

Example of IndexError

```python
lst = [10, 20, 30]

print(lst[5])
```

Q6. Best practices for exception handling in Python

Some important best practices are:

âœ” 1. Catch specific exceptions

Do not use a general except unnecessarily.

```python
try:
    x = int("abc")
except ValueError:
    print("Invalid number")
```

2. Do not hide exceptions

Avoid this:

```python
try:
    risky_code()
except:
    pass
```

3. Use finally for cleanup

```python
try:
    f = open("file.txt")
finally:
    f.close()
```

4. Use custom exceptions when needed

```python
class InvalidAgeError(Exception):
    pass
```

5. Do not use exceptions for normal logic

Use if conditions instead of forcing errors.

6. Give meaningful error messages

```python
raise ValueError("Age must be a positive number")
```

7. Keep try block small

Only put the code that may fail inside try.
